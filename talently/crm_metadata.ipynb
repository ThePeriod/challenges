```python
import pandas as pd
```

## _CRM_ Data Loading

```python
raw = pd.read_csv("./data/crm_data (1).csv")
null_treshold = 0.7
raw
```

## Data Validation

```python
#df_crm[df_crm.columns[0]].describe
metadata = pd.DataFrame({'Column Names' : [],
                         'Size' : [],
                         'Nulls' : [],
                         'Null %' : [],
                         'Unicos' : []})
```

```python
# Fill Columns field with Column names
metadata['Column Names'] = raw.columns
# Fill Size field with Column size
# Fill Nulls field with Column Nulls
for i in range(raw.columns.size):
    metadata['Size'][i] = (raw[raw.columns[i]].size)
    metadata['Nulls'][i] = (raw[raw.columns[i]].isnull().sum())
    metadata['Null %'][i] = (raw[raw.columns[i]].isnull().sum()/raw[raw.columns[i]].size).round(2)
    metadata['Unicos'][i] = raw[raw.columns[i]].nunique()
    if raw[raw.columns[i]].nunique() + raw[raw.columns[i]].isnull().sum() == raw[raw.columns[i]].size :
        metadata['Duplicados'] = False
    else:
        metadata['Duplicados'] = True
valids = metadata[metadata['Null %']<null_treshold].reset_index()
valids
```

```python
for i in valids['Column Names']:
    print(i)
    print(raw[i].unique())
    print(" ")
```
